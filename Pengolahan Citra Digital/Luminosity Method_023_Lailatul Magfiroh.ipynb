```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
from skimage import data
image = cv2.imread("D:\Fotokuu\Photograph\IMG_20200126_174905_1.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
x,y,z=image.shape
```

```python
# Gray = 0.21 R + 0.71 G + 0.07 B
W = [0.2,0.5,0.3] # weights
W_mean = np.tensordot(image,W, axes=((-1,-1)))[...,None]
image[:] = W_mean.astype(image.dtype)
```

```python
plt.figure(figsize=(10,20))
plt.imshow(image)
```
