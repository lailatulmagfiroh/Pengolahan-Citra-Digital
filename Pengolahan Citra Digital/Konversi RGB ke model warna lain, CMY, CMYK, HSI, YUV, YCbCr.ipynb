```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Buat program untuk melakukan proses:
# Konversi RGB ke model warna lain, CMY, CMYK, HSI, YUV, YCbCr
```

```python
image = cv2.imread("D:\Fotokuu\Photograph\IMG_20200126_174905_1.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
# RGB to CMY
```

```python
# Menghitung CMY
cmy = 255 - image

# Menampilkan gambar asli dan gambar dalam mode warna CMY
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].imshow(image)
ax[0].set_title('RGB Image')

ax[1].imshow(cmy)
ax[1].set_title('CMY Image')

plt.show()
```

```python
# RGB to CMYK
```

```python
cmy = cv2.bitwise_not(image)
cmy = cv2.cvtColor(cmy, cv2.COLOR_BGR2RGB)
cmy = cmy.astype(np.float32) / 255.0
c, m, y = cv2.split(cmy)
```

```python
k = 1 - np.maximum(np.maximum(c, m), y)
```

```python
c = (1 - c - k)/(1 - k)
m = (1 - m - k)/(1 - k)
y = (1 - y - k)/(1 - k)
```

```python
cmyk = cv2.merge((c, m, y, k))
```

```python
# Menampilkan gambar asli dan gambar dalam mode warna CMYK
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].imshow(image)
ax[0].set_title('RGB Image')

ax[1].imshow(cmyk)
ax[1].set_title('CMYK Image')

plt.show()
```

```python
smoothed_img = cv2.GaussianBlur(image, (5,5), 0)
```

```python
# Buat program untuk melakukan proses
# Sharpening citra berwarna
```

```python
# Definisi Sharpened
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
kernel = np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])
sharpened = cv2.filter2D(image, -1, kernel)
```

```python
# Menampilkan gambar asli dan gambar dalam mode sharpening
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].imshow(image)
ax[0].set_title('RGB Image')

ax[1].imshow(sharpened)
ax[1].set_title('Sharpening Image')

plt.show()
```
