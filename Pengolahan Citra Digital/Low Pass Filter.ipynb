```python
# Low Pass Filter
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("D:\EDIT\Folder\lowpict.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
kernel = np.array([[29, 10, 12, 13],
                 [34, 12, 13, 13],
                 [31, 10, 11, 12],
                 [30, 11, 14, 14],
                 [31, 12, 12, 11]])
```

```python
kernel = kernel/np.sum(kernel)
img2 = cv2.filter2D(image, -1, kernel)
```

```python
# Untuk plot image original
plt.imshow(image)
plt.title('Original')
plt.show()
# Untuk plot low pass image
plt.imshow(img2)
plt.title('Low Pass Filter')
plt.show()
```
